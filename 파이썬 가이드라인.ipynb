# 파이썬의 특징
## 1.휴머니즘 언어
### 프로그래밍은 인간의 생각을 컴퓨터에게 지시하는 것. 파이썬은 생각하는 방식을 그대로 표현하는 언어이다. 
### print("hello wolrd") 이렇게 직관적인 방식으로 생각하는대로 출력이 가능하다.
## 2.무료다
### 오픈 소스로 공개되어 있는 무료 소프트웨어이다. 그렇기 때문에 사용료 걱정 없이 언제 어디서든 파이썬을 다운로드 하여 사용이 가능하다.
## 3.간결하다.
### 소프트웨어 제작자가 의도적으로 간결하게 만들었다. 문법 또한 잘 정리되어 있기 때문에 다른 사람이 작업한 소스 코드를 이해하기도 수월하다. <hr>

```python
languages = ['python', 'perl', 'c', 'java']

for lang in languages:
    if lang in ['python', 'perl']:
        print("%6s need interpreter" % lang)
    elif lang in ['c', 'java']:
        print("%6s need compiler" % lang)
    else:
        print("should not reach here")
```
