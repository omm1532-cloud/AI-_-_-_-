```python
f = open("새 폴더//zz.txt",'r')
line = f.readline()
print(line)
while True:
    line = f.readline()
    break
f.close()
```

```python
f = open("새 폴더//zz.txt",'r')
lines = f.readlines()
for line in lines:
    print(line)
    f.close()
```
