# 문제2
## 텍스트 '20241014sunny' 에서 연도, 월, 일, 날씨를 각각 변수에 저장하여 출력해보자.
### text = '20241014sunny'

```python
#이 문제에선 문자열 인덱싱을 활용합니다. 파이썬은 숫자를 0부터 세기 때문에 '20241014sunny'라는 문장에서 연도 정보인 2024를 출력 하려면 숫자 0부터 시작한다.
#따라서 연도 정보는 year = text[0:4]로 입력후 연도를 출력하면 2024=year 이라는 값이 나오게 된다.
text = '20241014sunny'

year = text[0:4]
month = text[4:6]
day = text[6:8]
weather = text[8:13]

print(year+'=year')
print(month+'=month')
print(day+'=day')
print(weather+'=weather')
```

# 문제4
## 숫자를 오름차순과 내림차순으로 정렬해보자.
### a = [1,3,5,4,2]

```python
#나열 되어 있는 숫자를 오름차순으로 정렬하기 위하여 sort() 함수를 활용할 수 있다.
#그리고 내림차순으로 정렬하는 방법은 reverse() 함수를 활용하여 정렬된 숫자를 역순으로 바꿀수 있다.

a = [1,3,5,4,2]
a.sort();print(a)
a.reverse();print(a)
```

# 문제8
## 딕셔너리를 이용하여 'B'의 값을 구하여라. <br><br><hr>

```python
#딕셔너리는 키와 값으로 이루어진 자료형입니다. 딕셔너리를 활용하여 학생의 이름과 점수를 저장할 수 있습니다. 
#예를 들어 dic = {'A':90, 'B':80, 'C':70}로 딕셔너리를 설정하면 A라는 학생은 90점, B라는 학생은 80점, C라는 학생은 70점이라는 정보를 저장할 수 있습니다. 
#그리고 dic['B']로 입력하면 B라는 학생의 점수인 80이 출력됩니다.
dic = {'A':90, 'B':80, 'C':70}
print(dic['B'])
```

# 문제9
## 리스트에서 중복된 숫자를 제거하여 나열해보자.
### a = [1,1,1,2,2,3,3,3,4,4,5] <br><hr>

```python
#리스트에 중복되는 숫자는 1이 4개, 2가 2개, 3이 3개, 4가 2개, 5가 1개 있다. 
#리스트에서 중복되는 숫자를 제거하기 위해 del 함수를 활용하여 중복되는 숫자를 제거할 수 있다.
#먼저 del a[0],a[1],a[2],a[3],a[4]로 1이 4개 있는 숫자를 제거한다. 그리고 del a[3:4]로 2가 2개 있는 숫자 중 하나를 제거한다.
a = [1,1,1,2,2,3,3,3,4,4,5]
del a[0],a[1],a[2],a[3],a[4],a[3:4]
print(a)
```
