# 숫자형
- 숫자형(Number)은 숫자 형태로 이루어진 자료형이다. 정수, 실수, 8진수, 16진수 등이 여기에 해당한다.

- 다음 표는 파이썬에서 숫자를 사용하는 예이다.

- 정수: 1, 3, -5, -7, 0
- 실수: 1.1, 3.3, 4.5
- 8진수: 0o34, 0o25
- 16진수: 0x2A, 0xFF

```python
# 정수형은 말 그대로 양의 정수, 음의 정수, 0을 변수로 대입하는 형태다.

n = 11
print(n)
n = -35
print(n)
n = 0
print(n)
```

```python
# 실수형은 소수점이 포함된 숫자를 대입하는 형태다.

n = 3.14
print(n)
n = -1.5
print(n)
```

```python
# 8진수는 0o 또는 0O으로 시작하고, 16진수는 0x로 시작한다. 8진수나 16진수는 자주 사용하지 않으니 개념만 알아두자.

# 8진수
n = 0o177
print(n)

#16진수
n1 = 0x8ff
n2 = 0xABC
print(n2)
```

# 사칙연산

- 프로그래밍을 처음 접하더라도 사칙 연산(+, -, *, /)은 알고 있을 것이다. 파이썬도 계산기처럼 연산자를 사용해 사칙 연산을 수행한다.

```python
n = 30
m = 40
print(n + m)
print(n - m)
print(n*m)
print(n/m)
```

# a의 b제곱을 나타내는 ** 연산자
- ** 연산자는 a ** b처럼 사용하면 a의 b제곱값을 반환한다.

```python
a = 3
b = 4
a ** b
#= 81
print(a**b)
```

# 나눗셈 후 나머지를 반환하는 % 연산자
- %는 나눗셈의 나머지 값을 반환하는 연산자이다. 15을 2으로 나누면 나머지는 1.
- 2를 15로 나누면 나머지는 2.

```python
15 % 2
#= 1

2 % 15
#= 2

print(15 % 2)
print(2 % 15)
```

# 나눗셈 후 몫을 반환하는 // 연산자
- / 연산자를 사용하여 9 나누기 4를 하면 결과는 2.25가 된다.
- // 연산자를 사용하면 나눗셈 후 몫만 반환한다.

```python
9 / 4
#= 2.25
print(9 / 4)
```

```python
9 // 4
#= 2
print(9 // 4)
```

# 복합 연산자
- 복합 연산자는 산술 연산자(+, - 등)와 대입 연산자(=)를 합친 것이다.
- 먼저 x에 1을 대입한다.

```python
x = 1
x = x + 1  # a + 1로 계산된 값을 다시 a에 대입한다. a+= 1 은 a = a +1과 같다.
print(x)
```

- b += 1은 b = b + 1과 동일하다. 이처럼 +=를 복합 연산자라 부른다.
- b의 값을 1만큼 감소시키려면 -= 복합 연산자를 사용한다.

```python
b = 1
b -= 1  # a = a - 1과 같음
print(b)
```

- 복합 연산자에는 다음과 같은 것들이 있다.
- ex) +=, -=, *=, /=, //=, %=, **=
- 복합 연산자를 사용하면 코드를 더 간결하게 작성할 수 있다.

## 이스케이프 코드
- 문자열 예제에서 여러 줄의 문장을 처리할 때 역슬래시 문자와 소문자 n을 조합한 \n 이스케이프 코드를 사용했다. 
- 이스케이프(escape) 코드란 프로그래밍할 때 사용할 수 있도록 미리 정의해 둔 '문자 조합'이다. 
- 주로 출력물을 보기 좋게 정렬하는 용도로 사용한다. 몇 가지 이스케이프 코드를 정리하면 다음과 같다.

```python
# 이스케이프 코드의 종류는
# \n	문자열 안에서 줄을 바꿀 때 사용
#\t	문자열 사이에 탭 간격을 줄 때 사용
#\\	\를 그대로 표현할 때 사용
#\'	작은따옴표(')를 그대로 표현할 때 사용
#\"	큰따옴표(")를 그대로 표현할 때 사용
#\r	캐리지 리턴(줄 바꿈 문자, 커서를 현재 줄의 가장 앞으로 이동)
#\f	폼 피드(줄 바꿈 문자, 커서를 현재 줄의 다음 줄로 이동)
#\a	벨 소리(출력할 때 PC 스피커에서 '삑' 소리가 난다)
#\b	백 스페이스
#\000	널 문자
#이렇게 많지만 보통 활용 빈도가 높은 것은 \n, \t, \\, \',\" 이다.
print('hello\n \tpythone\\ \'world\"')
```

# 문자열 인덱싱과 슬라이싱
- 인덱싱(indexing)은 무엇인가를 '가리킨다', 슬라이싱(slicing)은 무엇인가를 '잘라 낸다'라는 의미다. 
- 이런 의미를 생각하면서 다음 내용을 살펴보자.



### 문자열 인덱싱

```python
x = "Life is too short, You need Python"
print(x[0])
print(x[33])
#문자 'L' 시작점은 0이고, 문자 'n' 마지막점은 33.
```

### 문자열 슬라이싱

```python
x = "Life is too short, You need Python"
y = x[0] + x[1] + x[2] + x[3]
print(y) #출력하고 싶은 부분을 슬라이싱 해서 변수로 설정 가능
y = x[0:5]
print(y) #좌표를 이용하여 설정도 가능하다 *시작은0부터 출력하고 싶은 문자 끝+1 까지 설정
```

### 슬라이싱 활용법

```python
# 'pithon'을 'pythone'으로 바꾸기
a = "Pithon"
print(a[:1])
print(a[2:])
print(a[:1] + 'y' + a[2:])
# 첫 번째 칸을 비워두면 0부터 두번째 칸을 비우면 끝까지 출력한다.
```

## 문자열 포맷팅
문자열에 변수나 값을 원하는 식으로 출력 하는 방식

- 1.숫자를 바로 대입
- 2.문자열을 바로 대입
- 3.어떤 변수를 통하여 값을 대입


## 문자열 포맷팅 코드
- %s	문자열(String)
- %c	문자 1개(character)
- %d	정수(Integer)
- %f	부동소수(floating-point)
- %o	8진수
- %x	16진수
- %%	Literal % (문자 % 자체)


```python
# 문자열 포맷 코드(d%)에 숫자를 바로 대입
print('시험 점수는 %d 입니다.' %70)

# 문자열 포맷 코드(s%)에 문자열을 바로 대입
print('만나서 반갑습니다. 오늘은 %s입니다.' %'목요일')

# 문자열 포맷 코드에 변수를 대입하는 방법
n = 5
print('강의 수심은 %d 미터입니다.' %n)

# 2개 이상의 문자열 포맷 코드 %(x, y)를 사용
x = '사람'
y = 10
print('여기 %s가 %d명 있습니다.' %(x, y))

# %d는 %s로 써도 무방함
x = '오리'
y = 30
print('강가에 %s들 %s마리가 지나갑니다.' %(x, y))
```

## 사용자 입력 함수 INPUT
사용자에게 문자열을 입력받아 변수에 저장하는 함수.
- 변수 = input('문자열을 입력하세요')
- 문자열 외의 다른 자료형을 사용하기 위해 자료형 변환을 사용해야함.

```python
# int = 정수
# float = 실수
x = int(input('정수를 입력하는 함수입니다.'))
y = float(input('실수를 입력하는 함수입니다.'))
print(x+y)
```

```python
# 자료형 변환, 변활할 변수(또는 값)을 사용하고 싶은 자료형으로 감싼다.

a = input('문자열을 입력해주세요: ')         # 사용자에게 값을 입력받아서 문자열으로 저장
b = int(input('정수를 입력해주세요: '))      # 사용자에게 값을 입력받아서 정수형으로 저장
c = float(input('실수를 입력해주세요: '))    # 사용자에게 값을 입력받아서 실수형으로 저장
print(a, b+c)
```

## 리스트 인덱싱
- 리스트도 문자열처럼 인덱싱과 슬라이싱이 가능하다.
-  *문자열이나 리스트를 리스트 내부에 사용하면 하나의 요소로 취급*


```python
a = [1, "그쪽은", -5, "위험합니다"]
a[0] = 1
a[1] = "그쪽은"
a[2] = -5
a[3] = "위험합니다"
print("a[0]: "+ str(a[0]))        # 정수 1을 문자열 "1"로 자료형 변환
print("a[1]: "+ a[1])
print("a[2]: "+ str(a[2]))
print("a[3]: "+ a[3])
```

## 리스트 슬라이싱
 - 단 리스트가 중복 되었을 때에는, 인덱싱한 후 슬라이싱을 해야한다.

```python
b = ['가', '나', '다', ['study', 'is', 'nice', ['x', 'y', 'z']]]

# 가, 나, 다를 가져오고 싶으면
print(b[0:3])

# ['study', 'is', 'nice'] 을 가져오고 싶으면
print(b[3][0:3])

# ['x', 'y', 'z']를 가져오고 싶으면
print(b[3][3])
```

## 리스트 뎃셈과 곱셈
리스트 덧셈 : 더하는 리스트가 하나로 합침

리스트 곱셈 : 곱하는 수만큼 리스트가 반복함

```python
a = [12, 21, 33]
b = [7, 10, 8]

# 리스트의 덧셈
c = a + b     # a+b는 a리스트와 b리스트가 합쳐져 변수 c에 저장됨
print(c)

# 리스트의 곱셈
print(a*3)
print(b*4)
```

## 리스트 관련 함수모음
 - LEN() : 리스트의 길이 측정함
 - APPEND() : 리스트에 요소를 추가함
 - SORT() : 리스트를 정렬함
 - REVERSE() : 리스트 순서를 뒤집기
 - INDEX() : 요소 위치 찾기
 - INSERT() : 요소 삽입하기
 - REMOVE() : 요소 제거하기
 - POP() : 요소 추출하기
 - COUNT() : 요소 개수 세기

### LEN함수
리스트의 길이를 측정합니다.

```python
a = [1, 2, 3, ['go', 'to', 'honme', ['q', 'w', 'e']]]

print(len(a))     # ['go', 'to', 'home', ['q', 'w', 'e']] 1개의 요소로 취급.
print(len(a[3]))  # ['q', 'w', 'e'] -> 1개의 요소로 취급.
print(len(a[3][0]))   # 'python'
```

### 리스트 요소 수정과 삭제및 추가
수정 : 요소에 직접 접근하여 수정

삭제 : del 사용

추가 : append() 함수를 사용

```python
a = [1, 2, 3, 4, 4, 4]
print(a)

a[5] = 7      # 5번 인덱스를 7로 변경
print(a)

del a[0]      # 리스트 첫번쨰 요소(0번 인덱스) 삭제
print(a)

del a[2:]     # 리스트 2번인덱스부터 끝까지 삭제
print(a)

a.append(4)
print(a)   # 리스트의 추가는 리스트의 맨 뒤에 이루어짐.
a.append([5,6,7,8])
print(a)  # append 명령어로는 한 번에 하나의 요소만 추가가 가능함.
```

### SORT 함수
리스트를 오름차순 정렬

```python
x = [2, 54, 7, 66, 31, 43, 11]
print(x)

x.sort()
print(x)

x = ['ze', 'ai', 'be', 'ei', 'le', 'ki']
print(x)

x.sort()
print(x)
```

### REVERSE 함수
리시트를 뒤집는다. 첫번째 요소부터 마지막 요소까지

```python
x = [2, 5, 7, 6, 3, 4, 1]
y = ['z', 'a', 'b', 'e', 'l', 'k']

print(x)
x.reverse()
print(x)

print(y)
y.reverse()
print(y) 
```

### INDEX 함수
원하는 요소의 위치를 찾는 함수
 - 리스트 내부에 같은 값이 존재할 때, 가장 앞에 있는 요소의 위치를 반환
 - 리스트 내에 없는 값을 찾을 시 에러가 발생한다.

```python
a = [1,2,3,4,5,1]
print(a.index(3))    # 3요소의 위치인, 2번 인덱스 반환
print(a.index(1))    # 1요소는 0, 5번 인덱스에 존재하지만, 가장 앞에 있는 0번 인덱스위치를 반환.
# 요소 위치가 존재 하지 않을 경우 에러 발생 ex) print(a.index(10))
```

### INSERT 함수
리스트에 원하는 위치에 요소를 삽입

```python
a = [1,2,3,4,10]
print(a)

a.insert(2, 10)
print(a)
```

### REMOVE 함수
리스트의 원하는 요소를 제거

중복된 값이 있으면 첫 번째 값만 제거한다.

```python
x = [1,2,3,1,4,5]
print(x)

x.remove(1)
print(x)

x.remove(3)
print(x)
```

### POP 함수
리스트의 원하는 위치에 있는 요소를 추출함

```python
x = [9,2,7,4,6]
print(x)

y = x.pop(2)      # pop함수를 통해 2번인덱스(3)을 추출하여 변수 b에 저장.
print(x)
print(y)
```

### COUNT 함수
리스트에 있는 원하는 요소의 개수를 세는 함수

```python
xyz = [12,3,4,5,1,2,2,45,2,1,3,4,2]
print(xyz)

print(xyz.count(3)) #3은 2개 존재 따라서 2출력
```

# 튜플 자료형(tuple)
리스트처럼 모든 자료형을 담을 수 있는 자료형.

튜플은 한 번 생성하면 값의 변경, 삭제가 불가능.

```python
# 모든 자료형을 담을 수 있는 자료형.
s1 = (1, 2, 3, ('python', 'is', 'good', ('s', 'b', 'c')))
print(s1)

# 튜플 내부에 요소가 하나만 존재할 때 맨 뒤에 ,를 붙인다.
s2 = (2,)
print(s2)

#튜플 생성시 괄호()를 생략해도 상관없다.
s3 = 1, 2, 3
print(s3)

# 1. 튜플 인덱싱
# 2. 튜플 슬라이싱
# 3. 튜플 더하기
# 4. 튜플 곱하기
# 5. LEN() 함수
# 위와 같은 기능을 사용할 수 있다.
```
